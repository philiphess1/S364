### Question 11
Create 2 top to bottom charts that displays the scatterplots of CRIM and MEDV. the one on the top should be houses with CHAS=0 and the one at the bottom CHAS=1. Use Boolean indexing to get 2 dataframes that contain houses with CHAS=0 and CHAS=1 respectively before plot the charts. Add x and y axis labels and legend like the output. See the accompanied output file for the Q11 resulting chart. Make sure your work matches it.

```python
...
### Your code starts 
def scatter_crim_medv_chas(housing_df):
    chas_0 = housing_df[housing_df['CHAS'] == 0]
    chas_1 = housing_df[housing_df['CHAS'] == 1]

    plt.figure(figsize=(8, 12))
    plt.subplot(2, 1, 1)
    plt.scatter(chas_0['CRIM'], chas_0['MEDV'], alpha=0.5, label='CHAS=0', color='steelblue')
    plt.title('Scatterplot of CRIM and MEDV (CHAS=0)')
    plt.xlabel('Crime Rate')
    plt.ylabel('MEDV')
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.scatter(chas_1['CRIM'], chas_1['MEDV'], alpha=0.5, label='CHAS=1', color='steelblue')
    plt.title('Scatterplot of CRIM and MEDV (CHAS=1)')
    plt.xlabel('Crime Rate')
    plt.ylabel('MEDV')
    plt.legend(loc = 'lower right')
    plt.tight_layout()
    plt.show()

scatter_crim_medv_chas(housing_df)
    
### Your code ends
```
